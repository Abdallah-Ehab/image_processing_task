```python
import matplotlib.pyplot as plt
import cv2
```

```python
original_image = cv2.imread("spider_man.jpg")
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
plt.imshow(original_image_rgb)
```

```python
original_image_rgb.shape
```

```python
def draw_hist(image,title):
    colors = ('b','g','r')
    for i,col in enumerate(colors):
        hist = cv2.calcHist([image],[i],None,[256],[0,256])
        plt.plot(hist,color=col,label=col+"channel")
        plt.xlim([0,256])
    plt.title(title)
    plt.show()
```

```python
draw_hist(original_image_rgb,"original_histogram")
```

```python
def adjust_image_and_show(contrast,brightness,image):
    new_image = cv2.convertScaleAbs(image, alpha=contrast, beta=brightness)
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title('original_image')
    plt.subplot(1, 2, 2)
    plt.imshow(new_image)
    plt.title('new_image')
    return new_image
    
```

```python
new_image = adjust_image_and_show(3,-50,original_image_rgb)
```

```python
draw_hist(new_image,'after increasing contrast and decreasing brightness')
```
